```python
total = []
for i in range(1, 18):
    img_data = plt.imread(f"golgi-movie_robert_frame_{i:>02}_gray.jpg")
    total.append(img_data.sum())
print(total)
```

```python
import numpy as np #can do maths stuff with numpy
total_array = np.array(total)
max_value = total_array.max()
percentage_array = total_array/max_value
```

```python
plt.figure(figsize=(17, 5)) #figsize sets the size of the fig. its a tuple based on height and width in inches
plt.plot([i for i in range(1, 18)], percentage_array, marker='o', linestyle='dashed') #use list comprehension
plt.xlabel('Frame Number')
plt.ylabel('Total fluorescence')
plt.xticks(range(1, 18)) #the 'tick' values for x axis

plt.show()
```
